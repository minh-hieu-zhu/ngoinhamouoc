```python
def hienthisanpham():
    print("---------------------------------------\n\n")
    for i in D_sanpham:
        print(i +": " + D_sanpham[i])
    print("\n\n---------------------------------------")
    
def kt_format(id_sp):
    if len(id_sp) != 4:
        return False
    if id_sp[:2] != "SP":
        return False
    if not id_sp[2:].isnumeric():
        return False
    return True

def themsanpham():  
    id_sp = input("Nhập id sản phẩm:")
    while not kt_format(id_sp) or id_sp in D_sanpham:
        if id_sp == "0":
            return
        print("Bạn vui lòng nhập đúng format SPXX và chắc chắn là id này chưa tồn tại trong danh sách!")
        id_sp = input("Nhập id sản phẩm:")
    name = input("Nhập tên sản phẩm:")
    D_sanpham[id_sp] = name

def suatensanpham():
    id_sp = input("Nhập id sản phẩm:")
    while id_sp not in D_sanpham:
        if id_sp == "0":
            return
        print("Bạn vui lòng nhập id có tồn tại trong danh sách!")
        id_sp = input("Nhập id sản phẩm:")
    name = input("Nhập tên sản phẩm:")
    D_sanpham[id_sp] = name

def xoasanpham():
    id_sp = input("Nhập id sản phẩm:")
    while id_sp not in D_sanpham:
        if id_sp == "0":
            return
        print("Bạn vui lòng nhập id có tồn tại trong danh sách!")
        id_sp = input("Nhập id sản phẩm:")
    D_sanpham.pop(id_sp)

while True:
    print("Nhập H nếu bạn muốn hiển thị danh sách sản phẩm")
    print("Nhập T nếu bạn muốn thêm sản phẩm")
    print("Nhập S nếu bạn muốn sửa sản phẩm")
    print("Nhập X nếu bạn muốn xóa sản phẩm")
    print("Nhập 0 nếu bạn muốn dừng chương trình")
    command = input("Lệnh của bạn:")
    if command == "H":
        hienthisanpham()
    elif command == "T":
        themsanpham()
    elif command == "S":
        suatensanpham()
    elif command == "X":
        xoasanpham()
    elif command == "0":
        break
```
