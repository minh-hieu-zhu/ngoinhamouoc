```python
import turtle

t = turtle.Turtle()

t.speed(10)
turtle.bgcolor('light blue');
for i in range(4):
    t.fd(100)
    t.rt(90)
t.lt(45)
t.fd(70)
t.rt(90)
t.fd(70)
t.lt(90)
t.fd(80)
t.rt(135)
t.fd(100)
t.rt(45)
t.fd(80)
t.bk(80)
t.lt(45)
t.bk(100)
t.rt(180)
t.lt(45)
t.fd(70)
t.lt(90)
t.fd(80)


t.penup()
t.fd(35)
t.lt(45)
t.fd(60)
t.pendown()

t.color("purple","purple")
t.begin_fill()
for i in range(2):
    t.fd(65)
    t.lt(90)
    t.fd(50)
    t.lt(90)
t.end_fill()

t.penup()
t.lt(90)
t.fd(250)
t.pendown()

t.color("brown","brown")
t.begin_fill()
for i in range(2):
    t.rt(90)
    t.fd(65)
    t.rt(90)
    t.fd(30)
t.end_fill()

t.color("green","green")
t.begin_fill()
t.fd(30)
for i in range(4):
    t.lt(120)
    t.fd(90)

t.lt(60)
t.fd(45)
for i in range(4):
    t.rt(120)
    t.fd(90)
    
t.lt(120)
t.fd(45)
for i in range(4):
    t.rt(120)
    t.fd(90)
t.end_fill()

t.penup()
t.lt(120)
t.rt(200)
t.fd(40)
t.pendown()

t.color("yellow","yellow")
t.begin_fill()
t.circle(50)
t.end_fill()
```
